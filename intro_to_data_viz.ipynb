```python
import pandas as pd
import plotly.express as px
```

```python
youtube_data=pd.read_csv('https://raw.githubusercontent.com/mawonuga121/randomdata/main/Top%20YouTube%20Channels%20Data%20.csv')
youtube_data
```

```python
fig=px.histogram(youtube_data, x= 'subscribers ', title= 'Subscriber Count')
fig.show()
```

```python
fig=px.pie(youtube_data, values= 'subscribers ', names='category ', title= 'YouTube Categories')
fig.show()
```

```python
fig=px.box(youtube_data, x= 'started ', title= 'Year Started')
fig.show()
```
